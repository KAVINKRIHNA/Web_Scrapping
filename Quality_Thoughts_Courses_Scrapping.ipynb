```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://qualitythought.in/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
Hot_Courses = soup.find('div', class_= "fusion-layout-column fusion_builder_column fusion_builder_column_1_1 fusion-builder-column-10 fusion-one-full fusion-column-first fusion-column-last 1_1")
```

```python
Hot_Courses
```

```python
Courses = Hot_Courses.find_all('h3')
Courses
```

```python
for  Course in Courses:
    for br in Course.find_all('br'):
        br.extract()
```

```python
Available_Courses = [titles.text.strip() for titles in Courses]
```

```python
Available_Courses
```

```python
df = pd.DataFrame(columns = Available_Courses)
```

```python
df
```

```python
row = Hot_Courses.find_all('p', class_ = 'Offer-price')
```

```python
row
```

```python
price = [titles.text.strip() for titles in row]
```

```python
price
```

```python
price_amounts = []
for p in row:
    price_text = p.get_text().strip()
    price_amount = price_text.split('JOIP:')[0].strip()
    price_amounts.append(price_amount)
```

```python
df.loc[0] = price_amounts
```

```python
print(df)
```
